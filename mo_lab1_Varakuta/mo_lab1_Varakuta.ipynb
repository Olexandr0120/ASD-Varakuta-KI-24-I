## Завдання 1:
Розв’язати задачу ЛП згідно з індивідуальним варіантом з Практичного заняття 1 (Графічний метод) з методички для практичних занять і самостійної роботи.

Знайти максимальне значення функції цілі:

$$
F = 5x_1 + 3x_2 \rightarrow \max
$$

за умов:

$$
\begin{cases}
6x_1 - 5x_2 \ge 1, \\
x_1 + 2x_2 \le 3, \\
4x_1 + 9x_2 \ge 1, \\
x_1 \ge 0,\; x_2 \ge 0.
\end{cases}
$$


```python
from scipy.optimize import linprog

# Цільова функція (мінімізуємо -F)
c = [-5, -3]

A_ub = [
    [-6, 5],
    [1, 2],
    [-4, -9]
]

b_ub = [-1, 3, -1]

bounds = [(0, None), (0, None)]

res1 = linprog(c, A_ub=A_ub, b_ub=b_ub, bounds=bounds, method="highs")

# Коректний вивід результатів
x1, x2 = res1.x
F_max = -res1.fun   # <-- тут прибираємо мінус ПРАВИЛЬНО

print("Статус:", res1.message)
print(f"x1 = {x1}")
print(f"x2 = {x2}")
print(f"Максимальне значення функції цілі F = {F_max}")

```

## Завдання 2  
### Чисельне розв’язання задачі ПЗ №2 (симплекс-метод) з використанням SciPy


```python
from scipy.optimize import linprog

# Цільова функція (мінімізація)
c2 = [45, 40, 40, 40]

# Обмеження типу <= (K1–K3)
A_ub_2 = [
    [3, 3, 4, 3],      # K1
    [0, 5, 4, 6],      # K2
    [15, 10, 12, 10]   # K3
]
b_ub_2 = [4, 5, 12]

# РІВНІСТЬ: вибір рівно одного вузла
A_eq_2 = [[1, 1, 1, 1]]
b_eq_2 = [1]

# Межі змінних
bounds_2 = [(0, None)] * 4

# Розв’язання симплекс-методом (HiGHS)
res2 = linprog(
    c=c2,
    A_ub=A_ub_2,
    b_ub=b_ub_2,
    A_eq=A_eq_2,
    b_eq=b_eq_2,
    bounds=bounds_2,
    method="highs"
)

res2
```

## Завдання 3  
### Розв’язання задачі ПЗ №1 з використанням пакета PuLP  

```python
from pulp import LpMaximize, LpProblem, LpVariable, LpStatus

model1 = LpProblem("PZ1_PuLP", LpMaximize)

x1 = LpVariable("x1", lowBound=0)
x2 = LpVariable("x2", lowBound=0)

model1 += 5*x1 + 3*x2

model1 += 6*x1 - 5*x2 >= 1
model1 += x1 + 2*x2 <= 3
model1 += 4*x1 + 9*x2 >= 1

model1.solve()

print("Status:", LpStatus[model1.status])
print("x1 =", x1.value())
print("x2 =", x2.value())
print("F =", model1.objective.value())
```

## Завдання 4  
### Розв’язання задачі ПЗ №2 з використанням пакета PuLP  
(симплекс-метод)


```python
from pulp import LpProblem, LpVariable, LpMinimize, LpStatus, value

model2 = LpProblem("PZ2_PuLP", LpMinimize)

x1 = LpVariable("x1", lowBound=0)
x2 = LpVariable("x2", lowBound=0)
x3 = LpVariable("x3", lowBound=0)
x4 = LpVariable("x4", lowBound=0)

model2 += 45*x1 + 40*x2 + 40*x3 + 40*x4

# Обмеження критеріїв
model2 += 3*x1 + 3*x2 + 4*x3 + 3*x4 <= 4
model2 += 5*x2 + 4*x3 + 6*x4 <= 5
model2 += 15*x1 + 10*x2 + 12*x3 + 10*x4 <= 12

# КЛЮЧОВЕ ОБМЕЖЕННЯ
model2 += x1 + x2 + x3 + x4 == 1

model2.solve()

print("Status:", LpStatus[model2.status])
print("x1 =", value(x1))
print("x2 =", value(x2))
print("x3 =", value(x3))
print("x4 =", value(x4))
print("F  =", value(model2.objective))
```

## Контрольні запитання
### 1. Яким чином працює симплекс-метод при розв’язанні задачі лінійного програмування?

Симплекс-метод — це ітеративний чисельний метод розв’язання задач лінійного програмування, який полягає у послідовному переході між вершинами багатогранника допустимих розв’язків з метою покращення значення цільової функції.

Загальний вигляд задачі лінійного програмування:
$$
\text{оптимізувати } F(\mathbf{x}) = \mathbf{c}^T \mathbf{x}
$$
за умов:
$$
\mathbf{A}\mathbf{x} \le \mathbf{b}, \quad \mathbf{x} \ge 0.
$$

Множина допустимих розв’язків є опуклим багатогранником. Симплекс-метод використовує властивість, що оптимальне значення цільової функції досягається в одній з вершин цього багатогранника.

### 2. Які основні кроки алгоритму симплекс-методу?

Основні етапи симплекс-методу:

1. Приведення задачі до стандартної форми:
$$
\max F = \mathbf{c}^T \mathbf{x}, \quad \mathbf{Ax} = \mathbf{b}, \quad \mathbf{x} \ge 0.
$$

2. Побудова початкового базисного допустимого розв’язку.

3. Обчислення оцінкових коефіцієнтів для визначення можливості покращення цільової функції.

4. Виконання симплекс-переходу, під час якого одна змінна входить у базис, а інша виходить з нього.

5. Перевірка умов завершення алгоритму.

### 3. Які недоліки має інструментарій бібліотеки SciPy у порівнянні з PuLP?

Бібліотека SciPy (функція `linprog`) має такі недоліки:

- підтримує лише задачі мінімізації;
- не дозволяє безпосередньо задавати обмеження типу
$$
\ge
$$
— їх потрібно перетворювати;
- не підтримує цілочисельні та бінарні змінні;
- модель задачі задається через матриці і вектори, що знижує читабельність.

**PuLP** дозволяє:
- напряму задавати максимізацію і мінімізацію;
- використовувати цілочисельні та бінарні змінні;
- описувати модель у вигляді, близькому до математичного запису.
